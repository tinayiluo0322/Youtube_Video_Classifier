```python
import pandas as pd

# List of columns to select
selected_columns = ["title", "tags", "categoryId", "description"]

# Dictionary to count categoryId occurrences
category_count = {}

# Set a chunk size
chunk_size = 150000

# First Pass: Count occurrences of each categoryId
for chunk in pd.read_csv(
    "US_youtube_trending_data.csv",
    chunksize=chunk_size,
    usecols=["categoryId"],
    low_memory=False,
):
    for categoryId in chunk["categoryId"]:
        if categoryId in category_count:
            category_count[categoryId] += 1
        else:
            category_count[categoryId] = 1

# Identify the top 4 most popular categoryIds
top_categories = sorted(category_count, key=category_count.get, reverse=True)[:4]

top_categories
```

```python
# Create an empty list to store filtered dataframes
filtered_data_list_2 = []

# Second Pass: Filter the data
for chunk in pd.read_csv(
    "US_youtube_trending_data.csv",
    chunksize=chunk_size,
    usecols=selected_columns,
    low_memory=False,
):
    # Filter the chunk for the top categories
    filtered_chunk = chunk[chunk["categoryId"].isin(top_categories)]
    # Append the filtered chunk to the list
    filtered_data_list_2.append(filtered_chunk)

# Combine all the filtered chunks into a single dataframe
final_df_2 = pd.concat(filtered_data_list_2, ignore_index=True)

final_df_2
```

```python
import re


def clean_and_tokenize(tag_string):
    # Remove non-alphanumeric characters and convert to lowercase
    cleaned_string = re.sub(r"[^a-zA-Z0-9\s|]", "", tag_string.lower())

    # Tokenize the string into words
    tokens = cleaned_string.split("|")

    # Remove leading and trailing whitespaces from each token
    tokens = [token.strip() for token in tokens]

    return tokens
```

```python
# Apply the cleaning and tokenization function to the 'Tags' column
final_df["tokenized_tags"] = final_df["tags"].apply(clean_and_tokenize)

# Display the DataFrame with the tokenized tags
final_df["tokenized_tags"]
```

```python
# Select only 'categoryId' and 'tokenized_tags' columns
new_df = final_df[["categoryId", "tokenized_tags"]]

# Display the new DataFrame
new_df
```

```python
# Convert lists to strings, replace commas with spaces, and optionally convert back to lists
new_df["tokenized_tags"] = new_df["tokenized_tags"].apply(
    lambda x: " ".join(x) if isinstance(x, list) else x
)

# Display the updated DataFrame
new_df
```

```python
# Filter the DataFrame for each category ID
df_category_10 = new_df[new_df["categoryId"] == 10]
df_category_17 = new_df[new_df["categoryId"] == 17]
df_category_20 = new_df[new_df["categoryId"] == 20]
df_category_24 = new_df[new_df["categoryId"] == 24]
```

```python
df_category_10.head()
```

```python
# Create a big list containing lists of tokens from each row in 'tokenized_tags' for category 10
big_list_of_tokenized_tags_10 = [
    row.split(" ") for row in df_category_10["tokenized_tags"]
]

# Print the big list
print(big_list_of_tokenized_tags_10[:3])
```

```python
# Specify the file path
file_path = "category10.txt"

# Open the file in write mode
with open(file_path, "w") as file:
    # Iterate through the list of lists
    for sublist in big_list_of_tokenized_tags_10:
        # Convert each element to a string and join them with a delimiter (e.g., comma)
        line = ",".join(map(str, sublist))

        # Write the line to the file
        file.write(line + "\n")
```
